```python
# Normalization
Age=data['Age']
minimum=data['Age'].min()
maximum=data['Age'].max()
```

```python
for i in Age:
    x=i-minimum/maximum-minimum
    print(x)
```

```python
# Normalization
Estsalary=data['EstimatedSalary']
minimum=data['EstimatedSalary'].min()
maximum=data['EstimatedSalary'].max()
```

```python
for i in Estsalary:
    x=i-minimum/maximum-minimum
    print(x)
```

```python
mean=Age.mean()
import statistics
std=statistics.stdev(Age)
```

```python
# Standardization
for i in Age:
    x=i-mean/std
    print(x)
```

```python
mean=Estsalary.mean()
import statistics
std=statistics.stdev(Estsalary)
```

```python
# Standardization
for i in Estsalary:
    x=i-mean/std
    print(x)
```
